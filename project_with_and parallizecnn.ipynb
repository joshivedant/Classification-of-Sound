```
lecture=np.load('lecture_mfcc.npy',allow_pickle=True)
lecture.shape
```

```
music=np.load('music_mfcc.npy',allow_pickle=True)
music.shape
```

```
silence=np.load('silence_mfcc.npy',allow_pickle=True)
len(silence)
```

```
v_lecture=np.array([1,0,0])
v_kirtan=np.array([0,1,0])
v_silence=np.array([0,0,1])
Y=v_lecture
for x in range(1,len(lecture)):
    Y=np.vstack((Y,v_lecture))
for x in range(len(music)):
    Y=np.vstack((Y,v_kirtan))
for x in range(len(silence)):
    Y=np.vstack((Y,v_silence))
Y
```

```
X=np.concatenate((lecture,music,silence),axis=0)
X.shape
```

```
music.shape
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.01)
```

```
X_train.shape
```

```
from numpy import asarray
from keras.models import Sequential,Input,Model
from keras.layers import Conv1D,Dense,MaxPool1D,Flatten,Dropout,Activation
x = Input(shape=(216,20))
y=Conv1D(16,(3),activation='relu',input_shape=(216,20))(x)
y=Dense(8, activation='relu')(y)
#y=Conv1D(32,(3),activation='relu')(y)
#y=Conv1D(64,(3),activation='relu')(y)
#y=Conv1D(128,(3),activation='relu')(y)
y=MaxPool1D(2)(y)
y=Dropout(0.5)(y)
y=Flatten()(y)
#y=Dense(64, activation='relu')(y)
y=Dense(3, activation='softmax')(y)
model = Model(inputs=x, outputs=y)
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
```
