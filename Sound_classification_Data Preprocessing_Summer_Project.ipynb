audio was chunked in 5 second windows

```python
i=0
from pydub import AudioSegment #for manipulating(splitting) audio files
sound_file = AudioSegment.from_file("Silence (1).mp3")
# milliseconds in the sound track
for x in range(0,len(sound_file),5000):
    new_file = sound_file[x : x+5000]
    new_file.export("silence"+ str(i) +".mp3", format="mp3")
    i=i+1
k=i
#repeat the process for speech and music data
```

Importing librosa for MFCC extraction

```python
import librosa
import numpy as np
```

**Extracting MFCC Features**

```python

silence=[]
%cd ./My Drive/Sound Classification
for x in range(0,k):
    name="silence" + str(x) +".mp3"
    y1, sr1 = librosa.load(name)
    arr=librosa.feature.mfcc(y=y1, sr=sr1, n_mfcc=20).T
    silence.append(arr)
#same for speech and music
```
